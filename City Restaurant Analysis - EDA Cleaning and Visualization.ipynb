# City Restaurants EDA

#### In this project, we will clean, analyze and visualize a city restaurant data to find useful insights.

###### About dataset:

The dataset contains the following columns with over 50000 records: 
- url:
contains the url of the restaurant in the zomato website

- address:
contains the address of the restaurant in Bengaluru

- name:
contains the name of the restaurant

- online_order:
whether online ordering is available in the restaurant or not

- book_table:
table book option available or not

- rate:
contains the overall rating of the restaurant out of 5

- votes:
contains total number of rating for the restaurant as of the above mentioned date

- phone:
contains the phone number of the restaurant

- location:
contains the neighborhood in which the restaurant is located

- rest_type:
restaurant type

- dish_liked:
dishes people liked in the restaurant

- cuisines:
food styles, separated by comma

- approx_cost(for two people):
contains the approximate cost for meal for two people

- reviews_list:
list of tuples containing reviews for the restaurant, each tuple consists of two values, rating and review by the customer

- menu_item:
contains list of menus available in the restaurant

- listed_in(type):
type of meal

- listed_in(city):
contains the neighborhood in which the restaurant is listed

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\mayan\Documents\datasets\zomato.csv")
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
# Dropping the columns that are not required for the analysis
df = df.drop(['url','address','phone','menu_item','dish_liked','reviews_list'],axis=1)
df.head()
```

```python
# Removing the duplicated rows present in the dataset.
df.drop_duplicates(inplace=True)
df.shape
```

```python
df.info()
```

```python
df['rate'].unique()
```

```python
# Creating a function to handle inconsistency of the values in rate column.
def handlerate(value):
    if (value=='NEW' or value=='-'):
        return np.nan
    else:
        value = str(value).split('/')
        value = value[0]
        return float(value)

df['rate'] = df['rate'].apply(handlerate)
df['rate'].head()
```

```python
df.isnull().sum()
```

```python
# Replacing NA values with mean value of the column.
df['rate'].fillna(df['rate'].mean(), inplace = True)
df['rate'].isnull().sum()
```

```python
# Removing all rows containing NA values.
df.dropna(inplace =True)
```

```python
# Renaming column names.
df.rename(columns = {'approx_cost(for two people)':'cost_for_two','listed_in(type)':'type'},inplace=True)
```

```python
df['location'].unique()
```

```python
df['listed_in(city)'].unique()
```

```python
# Dropping "listed_in(city)" because location column can be used instead.
df = df.drop(['listed_in(city)'],axis=1)
```

```python
df['cost_for_two'].unique()
```

```python
# Creating a function to handle "," present in the "cost_for_two" column, also changing the datatype to float.
def handlecomma(value):
    value = str(value)
    if ',' in value:
        value = value.replace(',','')
        return(value)
    else:
        return float(value)

df['cost_for_two'] = df['cost_for_two'].apply(handlecomma)
df['cost_for_two'].unique()
```

```python
df['rest_type'].value_counts()
```

```python
rest_type_count = df['rest_type'].value_counts()
```

```python
type_to_replace = rest_type_count[rest_type_count<1000].index
```

```python
# Changing the restaurant type where the count is less than 1000 to "others".
df['rest_type'] = df['rest_type'].apply(lambda x : 'others' if x in type_to_replace else x)
```

```python
df['rest_type'].value_counts()
```

```python
df.head()
```

```python
df['location'].value_counts(ascending=False)
```

```python
location_count = df['location'].value_counts()
```

```python
location_to_replace = location_count[location_count<300].index
```

```python
# Changing the location to "other" where the count is less than 300.
df['location'] = df['location'].apply(lambda x : 'other' if x in location_to_replace else x)
```

```python
df['location'].value_counts(ascending=False)
```

```python
df.head()
```

```python
df['cuisines'].value_counts(ascending = False)
```

```python
cuisine_count = df['cuisines'].value_counts()
cuisines_to_replace = cuisine_count[cuisine_count<100].index
```

```python
# Changing the cuisines to "others" where the count is less than 100.
df['cuisines'] = df['cuisines'].apply(lambda x : 'others' if x in cuisines_to_replace else x)
```

```python
df['type'].value_counts()
```

```python
df['location'].unique()
```

```python
# Count plot for various location.
plt.figure(figsize=(16,10))
ax = sns.countplot(x = 'location', data = df)
plt.xticks(rotation = 90)
```

```python
# Count plot for online order available.
plt.figure(figsize=(7,7))
sns.countplot(x='online_order',data=df)
```

```python
# Count plot for restaurants with book table option.
plt.figure(figsize=(7,7))
sns.countplot(x='book_table',data=df)
```

```python
# BoxPlot for online order vs rating.
plt.figure(figsize=(6,6))
sns.boxplot(x='online_order',y='rate',data=df)
```

```python
# BoxPlot for book table option vs rating.
plt.figure(figsize=(6,6))
sns.boxplot(x='book_table',y='rate',data=df)
```

```python
# Grouping the data by location and online order option and creating a pivot table for it.
df1 = df.groupby(['location','online_order'])['name'].count().reset_index()
df1 = pd.pivot_table(df1, values='name', index = ['location'], columns = ['online_order'], fill_value = 0, aggfunc = np.sum)
df1.columns.name = 'online_order'
df1
```

```python
# Plotting the location vs online order option.
df1.plot(kind='bar', figsize = (15,8))
```

```python
# Grouping the data by location and book table option and creating a pivot table for it.
df2 = df.groupby(['location','book_table'])['name'].count().reset_index()
df2 = pd.pivot_table(df2, values='name', index = ['location'], columns = ['book_table'], fill_value = 0, aggfunc = np.sum)
df2.columns.name = 'book_table'
df2
```

```python
# Plotting location vs book table option.
df2.plot(kind='bar', figsize=(18,10))
```

```python
# Plotting restaurants vs rating.
plt.figure(figsize=(10,7))
sns.boxplot(x='type',y='rate',data=df,palette='inferno')
plt.xticks(rotation = 45)
```

```python
# Grouping the data by location and restaurant type and creating a pivot table for it.
df3 = df.groupby(['location','type'])['name'].count().reset_index()
df3 = pd.pivot_table(df3, values='name', index = ['location'], columns = ['type'], fill_value = 0, aggfunc = np.sum)
df3.columns.name = 'type'
df3
```

```python
# Plotting the count with location and type of the restaurants
df3.plot(kind='bar', figsize=(36,9))
plt.xticks(rotation=45)
```

```python
# Grouping the data by location and votes and creating a pivot table for it.
df4 = df[['location','votes']]
df4.drop_duplicates()
df5 = df4.groupby(['location'])['votes'].sum().reset_index()
df5 = df5.sort_values('votes',ascending=False)
df5.head()
```

```python
# Plotting location vs rating.
plt.figure(figsize=(15,8))
sns.barplot(x='location',y='votes',data=df5)
plt.xticks(rotation=90)
```

```python
# Grouping the data by cuisines and votes and creating a pivot table for it.
df6 = df[['cuisines','votes']]
df6.drop_duplicates()
df7 = df6.groupby(['cuisines'])['votes'].sum().reset_index()
df7 = df7.sort_values('votes',ascending=False)
df7.head()
```

```python
# Excluding others category from the dataframe.
df7 =df7.iloc[1:,:]
df7
```

```python
# Plotting cuisines and votes.
plt.figure(figsize=(15,8))
sns.barplot(x='cuisines',y='votes',data=df7)
plt.xticks(rotation=90)
```
